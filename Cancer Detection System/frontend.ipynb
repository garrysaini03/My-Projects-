```python
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk
import cv2
import numpy as np
import joblib

# Load the pre-trained model
model = joblib.load("cancer_detection.joblib")

# Class label mapping
desc = {1: "Positive Cancer", 0: "No Cancer"}

# Upload and predict function
def upload_img():
    file_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.jpg *.png *.jpeg")])
    if not file_path:
        return

    try:
        # Load and preprocess image for prediction
        img_gray = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
        img_gray = cv2.resize(img_gray, (200, 200))
        img_array = img_gray.reshape(1, -1) / 255.0

        # Display image in GUI
        img_display = Image.open(file_path)
        img_display = img_display.resize((200, 200))
        img_tk = ImageTk.PhotoImage(img_display)

        img_label.config(image=img_tk)
        img_label.image = img_tk

        # Predict and show result
        prediction = model.predict(img_array)
        result_text = f"Prediction: {desc[prediction[0]]}"
        op.config(text=result_text, fg="green" if prediction[0] == 1 else "blue")
    except Exception as e:
        messagebox.showerror("Error", f"Failed to process image:\n{e}")

# GUI setup
app = tk.Tk()
app.title("Brain Tumor Classifier")
app.geometry("1080x1080")
app.configure(bg="#1ac080")

# Title Label
title = tk.Label(app, text="Brain Tumor Classifier", font=("Arial", 20, "bold"), bg="#F8FAFA", fg="black")
title.pack(pady=10)

# Frame for Image
frame = tk.Frame(app, bg="#FCFCFC")
frame.place(x=435, y=80, width=200, height=200)

# Upload Button
btn = tk.Button(app, text="Upload for breast cancer", font=("Arial", 15), bg="#F8F7F8", command=upload_img)
btn.place(x=100, y=400)

btn = tk.Button(app, text="Upload for blood cancer", font=("Arial", 15), bg="#F8F8FA", command=upload_img)
btn.place(x=400, y=400)

btn = tk.Button(app, text="Upload for bone cancer", font=("Arial", 15), bg="#F8F8FA", command=upload_img)
btn.place(x=700, y=400)

btn = tk.Button(app, text="Upload for lung cancer", font=("Arial", 15), bg="#F8F8FA", command=upload_img)
btn.place(x=200, y=500)
btn = tk.Button(app, text="Upload for malenoma", font=("Arial", 15), bg="#F8F8FA", command=upload_img)
btn.place(x=500, y=500)

# Image Display Label
img_label = tk.Label(frame)
img_label.pack()

# Output Prediction Label
op = tk.Label(app, font=("Arial", 15, "bold"), bg="#F8F8FA")
op.place(x=405, y=330)

# Start GUI
app.mainloop()
```
